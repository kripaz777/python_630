```python
fac = 1
n = int(input("Enter n = "))
if n >=0:
    for i in range(1,n+1):
        fac = fac*i

    print(fac)
else:
    print("The value of n should be n >= 0")
```

```python
n = int(input("Enter n = "))
for i in range(n):
    physics = int(input("Enter physics marks = "))
    chemistry = int(input("Enter chemistry marks = "))
    math = int(input("Enter  math marks = "))
    english = int(input("Enter english marks = "))
    nepali = int(input("Enter nepali marks = "))

    total = physics + chemistry + math + english + nepali

    per = total/5
    if per >= 80:
        grade = 'A'
    elif per >= 60:
        grade = 'B'
    elif per >= 45:
        grade = 'C'
    else:
        grade = 'F'

    print("The total marks = ",total)
    print("The per = ",per)
    print("The grade = ",grade)
```

```python
# WAp to create a billing system
# ---name,price,quantity,total,grand total
info = ""
grand_total = 0
n = int(input("Enter n = "))
for i in range(n):
    name = input("Enter name = ")
    price = int(input("Enter price = "))
    quantity = float(input("Enter quantity = "))
    total = price*quantity
    data = name + " " + str(price) + " " + str(quantity) + " " + str(total) + "\n"
    info = info + data
    grand_total = grand_total + total
print(grand_total)    
print(info)
```

```python
n = int(input("Enter n = "))
if n >=2:
    for i in range(2,n):
        if n%i==0:
            print("The no is composite.")
            break
    else:
        print("The no is prime.")
else:
    print("The no should be n >= 2")
```
